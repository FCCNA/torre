```python
def cerchio(x0, y0, R):
    X = np.linspace(x0-R, x0+R, 100)
    Yu = np.sqrt(R**2-(X-x0)**2)+y0
    Yd = -np.sqrt(R**2-(X-x0)**2)+y0
    plt.plot(X, Yu, color='blue')
    plt.plot(X, Yd, color='blue')
```

```python
def geometry(a, bottom, top):
    if bottom == True :
        a.scatter(wires0011_x, wires0011_z, color='gray', s=10)
        a.scatter(wires1223_x, wires1223_z, color='gray', s=10)
        a.axhline(y=wires0011_z[0], color='gray', linestyle='-')
        a.axhline(y=wires1223_z[0], color='gray', linestyle='-')
        
    if top == True :
        a.scatter(wires4859_x, wires4859_z, color='gray', s=10)
        a.scatter(wires6071_x, wires6071_z, color='gray', s=10)
        a.axhline(y=wires4859_z[0], color='gray', linestyle='-')
        a.axhline(y=wires6071_z[0], color='gray', linestyle='-')
    
    return
```

```python
import sympy

# Definizione dei simboli
x0=0
y0=0
r=1
x, y, x0, y0, r, m, q = sympy.symbols('x y x0 y0 r m q')

# Equazione della circonferenza
eq_circle = (x - x0)**2 + (y - y0)**2 - r**2

# Derivata parziale di y rispetto a x
dy_dx = sympy.diff(eq_circle, x) / sympy.diff(eq_circle, y)

# Sostituzione di y con m*x + q per trovare l'intercetta
eq_line = dy_dx*(x - x0) + y0 - m*x - q

# Risoluzione del sistema di equazioni lineari
sol = sympy.solve([eq_line, eq_circle], [m, q])

# Stampa del risultato
print(f"Coefficiente angolare: {sol[m]}")
print(f"Intercetta: {sol[q]}")
```

```python
import math

def tangents_to_circle(x0, y0, r, x):
    """Calcola le equazioni delle rette tangenti alla circonferenza di centro (x0, y0) e raggio r che passano per il punto x."""
    dx = x - x0
    dy = math.sqrt(r ** 2 - dx ** 2)
    m = -dx / dy
    q = x * (y0 - dy) / dy + y0
    return (m, q), (m, q - 2 * y0)

# Esempio di utilizzo
x0, y0, r = 0, 0, 2
x = 1
t1, t2 = tangents_to_circle(x0, y0, r, x)
print(f"Le rette tangenti che passano per il punto ({x}, {t1[0] * x + t1[1]}) sono: y = {t1[0]}x + {t1[1]} e y = {t2[0]}x + {t2[1]}")
```
